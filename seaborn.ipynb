```python
import seaborn as sns
```

Seaborn is a python visualization library based on matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics



```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
taxis = sns.load_dataset('taxis')
penguins = sns.load_dataset('penguins')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
```

```python
##Scatterplot
import matplotlib.pyplot as plt
sns.scatterplot(x ='total_bill', y = 'tip',data = tips , hue = 'day', size = 'size')
plt.title('Scatterplot of total_bill vs tips')
```

```python
##Line plot
sns.lineplot(x = "size" , y = 'total_bill', data = tips)
plt.title('Line plot ')
```

```python
#Bar Plot
sns.barplot( x = 'day', y = 'total_bill', data = tips)
plt.title('Bar Plot')
```

```python
sns.barplot( x= 'sex', y = 'tip', data = tips, palette='YlGnBu')
```

```python
## Boxplot
sns.boxplot(x ='day', y = 'total_bill', data = tips)
plt.title('Boxplot')
```

```python
sns.boxplot(x = 'day', y = 'tip',data = tips,hue = 'sex',palette = 'YlGnBu' )
```

```python
## Strip Chart
sns.stripplot(x = 'day', y = 'tip', data = tips , hue = 'sex', palette = 'YlGnBu')
```

```python
#Vilon Plot
sns.violinplot(x = 'day', y = 'total_bill',data = tips)
```

```python
#histograms
sns.histplot( tips["total_bill"], bins = 15, kde= True)#kde = Kernel Density Emulator
```

```python
# KDE plot
sns.kdeplot(tips['total_bill'],fill ='blue')
```

```python
#Joint plot
sns.jointplot(x = 'tip', y ='total_bill', data = tips, kind ='hex',cmap = 'YlGnBu')
```

```python
#Joint plot
sns.jointplot(x = 'tip', y ='total_bill', data = tips, kind ='kde', shade = True,cmap = 'YlGnBu')
```

```python
#Pairplot
sns.pairplot(tips, hue ='sex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(titanic)
plt.title("Titanic")
```

```python
sns.pairplot(titanic , hue ='pclass')
plt.title("Titanic")
```
